```python
import pandas as pd
import numpy as np

data = pd.read_csv(r'NSL_Regular_Season_Data.csv')
data = data[data["n"] == data["n"]].iloc[:,:17]

team = pd.DataFrame(index=data.sort_values(by="HomeTeam",ascending=True).HomeTeam.unique(),columns=['rating'])

#set base rating
team['rating'] = 1500

data
```

```python
k = 100
team['rating'] = 1500
ratings = pd.DataFrame()
ratings.insert(len(ratings),'initial',team)

#do multiple iterations, entering the new ratings as initial ratings
for iterations in range(10):
    #be able to test different k values
    #for k in range(k,k+1,5):
        for i in range(len(data)):
            home_team = data.loc[i,'HomeTeam']
            away_team = data.loc[i,'AwayTeam']
            home_rating = team.loc[home_team,'rating']
            away_rating = team.loc[away_team,'rating']
            if data.loc[i,'HomeScore']>data.loc[i,'AwayScore']:
                home_score = 1
                away_score = 0
            elif data.loc[i,'HomeScore']<data.loc[i,'AwayScore']:
                home_score = 0
                away_score = 1
            else:
                home_score = 0.5
                away_score = 0.5
            home_expected = 1/(1+10**((away_rating-home_rating)/400))
            away_expected = 1 - home_expected
            team.loc[data.loc[i,'HomeTeam'],'rating'] = home_rating + k*(home_score-home_expected)
            team.loc[data.loc[i,'AwayTeam'],'rating'] = away_rating + k*(away_score-away_expected)
        #ratings.insert(len(ratings.transpose()),'k='+str(k),team)
        ratings.insert(len(ratings.transpose()),'iteration='+str(iterations),team)

ratings

```

```python
ratings_sorted = ratings.sort_values('iteration=9',ascending=False)
ratings_sorted
```

```python
ratings_sorted.to_csv('elo_ranked_teams.csv')
```
