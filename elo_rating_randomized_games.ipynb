```python
import pandas as pd
import numpy as np

data = pd.read_csv(r'NSL_Regular_Season_Data.csv')
data = data[data["n"] == data["n"]].iloc[:,:17]

team = pd.DataFrame(index=data.sort_values(by="HomeTeam",ascending=True).HomeTeam.unique(),columns=['rating'])

#set base rating
team['rating'] = 1500

data
```

```python
k = 100
team['rating'] = 1500
ratings = pd.DataFrame()
ratings.insert(len(ratings),'initial',team)


#do multiple iterations, entering the new ratings as initial ratings
for random_iterations in range(10):
    randomized_data = data.sample(len(data))
    randomized_data = randomized_data.reset_index().iloc[:,1:]
    for iterations in range(5):
    #be able to test different k values
    #for k in range(k,k+1,5):
        for i in range(len(randomized_data)):
            home_team = randomized_data.loc[i,'HomeTeam']
            away_team = randomized_data.loc[i,'AwayTeam']
            home_rating = team.loc[home_team,'rating']
            away_rating = team.loc[away_team,'rating']
            if randomized_data.loc[i,'HomeScore']>randomized_data.loc[i,'AwayScore']:
                home_score = 1
                away_score = 0
            elif randomized_data.loc[i,'HomeScore']<randomized_data.loc[i,'AwayScore']:
                home_score = 0
                away_score = 1
            else:
                home_score = 0.5
                away_score = 0.5
            home_expected = 1/(1+10**((away_rating-home_rating)/400))
            away_expected = 1 - home_expected
            team.loc[randomized_data.loc[i,'HomeTeam'],'rating'] = home_rating + k*(home_score-home_expected)
            team.loc[randomized_data.loc[i,'AwayTeam'],'rating'] = away_rating + k*(away_score-away_expected)
        #ratings.insert(len(ratings.transpose()),'k='+str(k),team)
            
    ratings.insert(len(ratings.transpose()),'randomized_sample_'+str(random_iterations),team)

ratings

```

```python
scores = pd.DataFrame()
final_home_score = ratings.sort_values('iteration_home=9',ascending=False)['iteration_home=9'].reset_index().rename(columns={"index": "home_team"})
final_away_score = ratings.sort_values('iteration_away=9',ascending=False)['iteration_away=9'].reset_index().rename(columns={"index": "away_team"})
scores = pd.concat([final_home_score,final_away_score],axis=1)
scores
```

```python
scores.to_csv('elo_ranked_home_and_away.csv')
```
