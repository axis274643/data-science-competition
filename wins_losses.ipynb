```python
import pandas as pd
import numpy as np

data = pd.read_csv(r'NSL_Regular_Season_Data.csv')
data = data[data["n"] == data["n"]].iloc[:,:17]

data_home = data.sort_values(by="HomeTeam",ascending=True)
data_away = data.sort_values(by="AwayTeam",ascending=True)

data
```

```python
home_wins = data_home[data_home['HomeScore']>data_home['AwayScore']].pivot_table(index=['HomeTeam'],aggfunc='size')
away_losses = data_home[data_home['HomeScore']>data_home['AwayScore']].pivot_table(index=['AwayTeam'],aggfunc='size')
home_losses = data_home[data_home['HomeScore']<data_home['AwayScore']].pivot_table(index=['HomeTeam'],aggfunc='size')
away_wins = data_home[data_home['HomeScore']<data_home['AwayScore']].pivot_table(index=['AwayTeam'],aggfunc='size')
home_draws = data_home[data_home['HomeScore']==data_home['AwayScore']].pivot_table(index=['HomeTeam'],aggfunc='size')
away_draws = data_home[data_home['HomeScore']==data_home['AwayScore']].pivot_table(index=['AwayTeam'],aggfunc='size')
```

```python
games=pd.DataFrame(zip(home_wins,away_wins,home_wins+away_wins,home_losses,away_losses,home_losses+away_losses,home_draws,away_draws,home_draws+away_draws),index=home_wins.index,columns=['Home Wins','Away Wins','Total Wins','Home Losses','Away Losses','Total Losses','Home Draws','Away Draws','Total Draws'])
```

```python
games.to_csv('team_wins_losses_draws.csv')
```
